```python
url2 = 'https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue'
page_2 = requests.get(url2) 
```

```python
soup_2 = BeautifulSoup(page_2.text, 'html')
```

```python
soup_2.find('table')
```

```python
table = soup_2.find_all('table')[1]
```

```python
world_titles = table.find_all('th')
```

```python
print(world_titles)
```

```python
world_table_titles = [title.text.strip() for title in world_titles]
```

```python
print(world_table_titles)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = world_table_titles)
df
```

```python
column_data = table.find_all('tr')
```

```python
print(column_data)
```

```python
for values in column_data[1:]:
    row_data = values.find_all('td')
    individual_data = [data.text.strip() for data in row_data]
    
    length = len(df)
    df.loc[length] = individual_data
    
```

```python
df
```
